# Lists or Dynamic Arrays

### Properties:

- Heterogeneous: Yes
    - Can contain elements of different data types.
    - eg. `[1, "two", 3.0, True]`

- Ordered: Yes
    - Elements maintain their insertion order.
    - eg. `[1, 2, 3]` as in '1' was inserted at first hence comes before the other entries.

- Mutable: Yes
    - Elements can be changed, added, or removed after creation.
    - eg. `my_list = [1, 2, 3]; my_list[0] = 10` changes the first element to 10. my_list becomes `[10, 2, 3]`

- Indexed: Yes
    - Elements can be accessed via their index positions.
    - eg. `my_list[1]` accesses the second element of the list.
    - starting from `0` for the first element till the `n-1` for the `n`th element.
    - Python also supports negative indexing where `-1` refers to the last element, `-n` refers to the first element.

- Duplicates Allowed: Yes
    - Multiple identical elements can exist in the list.
    - eg. `[1, 2, 2, 3]` contains two '2's.

- Dynamic Size: Yes
    - Size can grow or shrink as elements are added or removed.
    - eg. `my_list.append(4)` adds an element to the end of the list.
    - No need to declare size beforehand as done in Java or C++ arrays.
    
    - Lists don't store objects directly; they store references to objects. This allows for dynamic resizing.
    - When the list exceeds its current capacity, a new larger block of memory is allocated and existing elements are copied over.

- Nested Structures: Yes
    - Lists can contain other lists or data structures as elements.
    - eg. `[1,2,"dev", (2,3), {"name":"Khushi", "surname":"Jani"}, [1,2,3]]` 


- Memory Allocation: Contiguous
    - Elements are stored in contiguous memory locations.
    - This allows for efficient access and iteration.

    What is Contiguous Memory Allocation?
    - In contiguous memory allocation, all elements of a data structure are stored in a single, continuous block of memory.
    - This means that the memory addresses of the elements are sequential, allowing for efficient access and iteration.

- Performance: Fast access and modification 
    - Accessing elements by index is O(1) time complexity.
    - Adding or removing elements can be O(n) in the worst case due to resizing, but average case is O(1) for appending.

# Timsort: The Adaptive Engine Under the Hood

<b>Timsort = Merge Sort + Insertion Sort</b>

The Strategy unfolds in two main phases:

1. <b>Run Detection and Preparation</b> : 
    - The array is scanned to identify "runs," which are sequences of consecutive elements that are already sorted (either ascending or descending).
    - If a run is found to be descending, it is reversed to make it ascending.
    - If a run is shorter than a predefined minimum length (commonly 32), Timsort extends it using Insertion Sort. This is because Insertion Sort is efficient for small datasets due to its low overhead.

2. <b>Merge Phase</b> :
    - Once the runs are identified and prepared, Timsort merges them using a process similar to Merge Sort.
    - The merging process takes advantage of the fact that the runs are already sorted, allowing for efficient merging.

### Why Timsort Excels
- **Adaptive**: Timsort adapts to the existing order in the data, making it faster for partially sorted arrays.
- **Stable**: It maintains the relative order of equal elements, which is important for certain applications.
- **Efficient**: With a worst-case time complexity of O(n log n) in the average and worst cases, and O(n) in the best case (when the array is already sorted), Timsort is efficient for a wide range of scenarios.
- **Memory Usage**: Timsort requires O(n) additional space for the temporary arrays used during the merge process, which is a trade-off for its speed and stability.

## Classic Sorting Algorithms

### O(n<sup>2</sup>)

#### Bubble Sort

Bubble sort is the simple sorting algorithm that iterates through the list, comaparing adjacent elements & swapping them if they are in the wrong order. This process is repeated until the list is sorted.

Best Time Complexity: `O(n)` - when the array is already sorted, only one pass is needed to confirm the order.

Average and Worst Time Complexity: `O(n`<sup>`2`</sup>`)` - when the array is in reverse order, every element needs to be compared with every other element.

````python
def bubble_sort(li):
    for c in range(len(li), 0, -1):
        index = 0
        while index + 1 < c:

            if li[index] > li[index + 1]:
                current = li[index]
                li[index] = li[index + 1]
                li[index + 1] = current

            index += 1

    return li


li = [5, 2, 4, 1, 2]
print(bubble_sort(li))
````

````python
# Optimised version of Bubble Sort
# It has nothing just it checks if the list is already sorted or not.

def bubble_sort(li):
    
    iterations = 0
    
    for c in range(len(li), 0, -1):
        noSwap = True   
        index = 0
        while index + 1 < c:

            iterations += 1

            if li[index] > li[index + 1]:
                noSwap = False
                current = li[index]
                li[index] = li[index + 1]
                li[index + 1] = current

            index += 1

        if noSwap:
            break

    return li, iterations

li = [3,2,1,4,5]
print(bubble_sort(li))
````

### Selection Sort

This algorithm divides the input list into two parts one sorted and other is unsorted. It repeatedly selects the smallest (or largest, depending upon the algorithm) element from the unsorted part and moves it to the end/beginning of the sorted part.

Best, Average and Worst Time Complexity: `O(n`<sup>`2`</sup>`)` - regardless of the initial order of elements, the algorithm always performs n*(n-1)/2 comparisons.

Space Complexity: `O(1)` - it only requires a constant amount of additional memory space for the temporary variable used for swapping.

````python
def selection_sort(li):
    size = len(li)
    for step in range(size):    
        min_index = step
        for i in range(step + 1, size):

            if li[min_index] > li[i]:
                min_index = i

        li[min_index], li[step] = li[step], li[min_index]

    return li 

li = [5, 1, 2, 4, 3]
selection_sort(li)
````

### Insertion Sort

Insertion sort builds the array using one element at a time.
Everytime the list is iterated it selects a minimum element and inserts it into the appropriate position somewhere in the middle of the original list and at the end of sorted list.

The elements greater then the current smallest element are shifted right or towards the end of the list.

Worst and Average Time Complexity: `O(n`<sup>`2`</sup>`)` - occurs when the array is sorted in reverse order, requiring each new element to be compared with all previously sorted elements.

Best Time Complexity: `O(n)` - occurs when the array is already sorted, requiring only one comparison per element.

<image src="../0_Ow0jhpLDTi4ZtKhU.png" alt="Insertion Sort" width="100%"/>

````python
def insertion_sort(li):
    size = len(li)
    for step in range(1, size):
        key = li[step]
        j = step - 1
        while j >= 0 and li[j] > key:
            li[j + 1] = li[j]
            j -= 1
        li[j + 1] = key
    return li

li = [85,12,59,45,72,51]
insertion_sort(li)
````

## O(n log n)

### Merge Sort

Classic Example of the Divide and Conquer Algorithm.

The algorithm recursively divides the array into two halves until the array is divided into single element arrays(which are automatically sorted arrays). Then this individual divided arrays are brought together using specific method.

Worst, Average, Best Time Complexity :- `O(n log n)`

Space Complexity :- `O(n)`

This is a crucial space-time trade-off. While it is less space-efficient than an in-place algorithm, its stability and reliable performance make it an excellent choice for scenarios where data can fit into memory (external sorting) or when preserving the relative order of equal elements is critical.

### How it works?

1. Everytime the array is splitted into two parts.
2. Two parts are sorted into ascending order and merged everytime the function is called recusively.
3. When only one element is left in the array it is returned as it is.

````python
# Merge Sort

def merge_sort(arr):

    if len(arr) == 1:
        return arr

    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    
    return mergeArraysSorting(left, right)


def mergeArraysSorting(arr1: list, arr2: list) -> list:
    merged_array = list()
    i = j = 0
    while i < len(arr1) and j < len(arr2):
        if arr1[i] < arr2[j]:
            merged_array.append(arr1[i])
            i += 1
        else:
            merged_array.append(arr2[j])
            j += 1

    merged_array.extend(arr1[i:])
    merged_array.extend(arr2[j:])

    return merged_array

merge_sort([0,9,4,2,5,7,10,1])
````

### In-place Merge-sort Algoritm

````python
def inPlace_merge_sort(arr: list, start: int, end: int):
    # Base case: if the segment has 1 or 0 elements
    if end - start <= 0:
        return

    if start < end:
        mid = start + (end - start) // 2

        # Sort first and second halves
        inPlace_merge_sort(arr, start, mid)
        inPlace_merge_sort(arr, mid + 1, end)

        # Merge the sorted halves
        merge_inPlace(arr, start, mid, end)

def merge_inPlace(arr, start, mid, end):
    start2 = mid + 1

    # If the direct merge is already sorted
    if arr[mid] <= arr[start2]:
        return

    # Two pointers to maintain start of both arrays
    while start <= mid and start2 <= end:
        
        # If element 1 is in right place
        if arr[start] <= arr[start2]:
            start += 1
        else:
            value = arr[start2]
            index = start2

            # Shift all elements between element 1 and element 2 right by 1
            while index != start:
                arr[index] = arr[index - 1]
                index -= 1
            
            arr[start] = value

            # Update all pointers
            start += 1
            mid += 1
            start2 += 1
````

### Quick Sort

Another Prominient Divide and Conquer Algorithm.

It works by selecting a "Pivot" element and partitioning the array around it. Such that all the elements smaller then pivot is set on the left of it and larger on the right of element. The algorithm is recursively applied on the array until the array is sorted.

Worst Time Complexity :- `O(n`<sup>`2`</sup>`)`<br>
This occurs when the pivot selection consistently results in unbalanced partitions, such as when a naive pivot strategy is applied to a pre-sorted or reverse-sorted array. This vulnerability can be mitigated by choosing better pivot strategies, such as selecting a random element or the median of three, which helps ensure that the average-case performance holds true more often in practice. A standard Quick Sort implementation is also not stable.

Best , Average Time Complexoty :- `O(log n)`

Space Complexity :- `O(log n)`

`NOTE:` Quick Sort & Merge Sort will be coded in recurssion.

#### DSA Problem solving patterns with methods:

Sorting as a whole is not a complete question; It is used to preprocess the data to use the more efficient DSA Algorithm such as:

1. <u>Two-Sum</u> :The problem of finding two numbers in an array that sum to a specific target value can be solved in O(n^2) time with a brute-force approach. However, if the array is first sorted in O(nlogn), the problem can then be solved in O(n) time using a two-pointer approach, where pointers from both ends of the array converge inward. The overall time complexity becomes O(nlogn), which is significantly more efficient for large datasets.

2. <u>Merge Intervals</u>: This problem involves merging overlapping intervals in a list. The first step is to sort the intervals based on their start times. This sorting operation simplifies the problem, allowing for a single linear pass through the sorted intervals to perform the merge operation.
