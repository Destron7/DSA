## String Mutability

A cornerstone concept in Python's string handling is their immutability. This means that once a string object has been created, its content cannot be altered. Any operation that appears to "modify" a string, such as concatenation or character replacement, does not change the original string in place. Instead, these operations always produce a new string object in memory, leaving the original string untouched. 

The slightest of modification to the string will result in a new string being created. The variable pointing to the address to the original string will now point to the new string. Resulting into more garbage collection overhead and potential memory inefficiencies. This should be considered as it increases memory usage and sometimes in case of loops also impacts performance.

### Efficient String Building and Performance Optimization



````python
s = ""
c = "a"
for i in range(M:= 10):
    s += c
    print(s)
````

````python
import time
import sys

def concatenate_with_plus(M):
    """Measures time and cumulative memory for the += method."""
    s = ''
    c = 'a'
    cumulative_size = 0
    start_time = time.time()
    
    for _ in range(M):
        s += c
        # We add the size of the newly created string to a running total.
        cumulative_size += sys.getsizeof(s)
        
    end_time = time.time()
    return end_time - start_time, s, cumulative_size

def concatenate_with_join(M):
    """Measures time and cumulative memory for the join method."""
    lst = []
    c = 'a'
    cumulative_size = 0
    start_time = time.time()
    
    for _ in range(M):
        lst.append(c)
        
    result = ''.join(lst)
    cumulative_size = sys.getsizeof(result) # Only one major string allocation
    end_time = time.time()
    
    return end_time - start_time, result, cumulative_size

# Set a number of iterations to demonstrate the difference clearly
M = 10000

print(f"Measuring performance for M = {M} iterations...\n")

# Measure for the original += code
time_plus, result_plus, cumulative_plus = concatenate_with_plus(M)

print("--- Original Code (s += c) ---")
print(f"Time taken: {time_plus:.6f} seconds")
print(f"Cumulative memory allocated (conceptual): {cumulative_plus} bytes")
print(f"Final string size: {sys.getsizeof(result_plus)} bytes\n")

# Measure for the efficient join method
time_join, result_join, final_size_join = concatenate_with_join(M)

print("--- Efficient Alternative (str.join) ---")
print(f"Time taken: {time_join:.6f} seconds")
print(f"Final string size: {final_size_join} bytes")
````

As strings are <b>immutable</b> in Python; each iteration of the for the loop in the function always allocates the memory for the every string made during the process till the desired resultant is achieved. Hence increasing the space complexity temendously in the manner of 1 + 2 + 3 + .... + M = `M * (M + 1) / 2`. Hence giving Space complexity of `O(N^2)`. So, for the longer strings `+=` becomes inefficient and can lead to performance bottlenecks.

While the `str.join()` method is more efficient because it calculates the total size of the final string in advance and allocates memory for it just once, rather than creating many intermediate strings. Once the memory is allocated the characters are copy pasted at there final locations.

## F-Strings

`f"This is the format for F-String!!"`

Generally F-Strings are used for formatting a string. They provide a way to embed expressions inside string literals, using curly braces `{}`. This makes it easier to create complex strings without needing to use concatenation or the `str.format()` method.

Formatting specifiers:

1. Decimal Places:<br>
        `f"{value:.2f}"`<br>
        for controlling the number of decimal places.

2. Alignment and Padding:<br>
        `f"{value:<10}"` for left-aligning within a 10-character field.<br>
        `f"{value:>10}"` for right-aligning within a 10-character field.<br>
        `f"{value:^10}"` for centering within a 10-character field.<br>
        `f"{value:0>10}"` for padding with zeros on the left to make it 10 characters wide.

3. Thousand Separators:<br>
        `f"{value:,}"` for including commas as thousand separators.

4. Using `f"{variable=}"` directly <b>displays both the variable's name and its value</b>.

````python
name = "Dev patel"
number = 10234.23508283942048
````

````python
f"{name:^20}"
````

````python
f"{number=:^20,.5f}"
````

### Q. Check if the string is a palindrome or not!?

````python
check_string = ""
i = 0
j = len(check_string) - 1
while i < j:
    if check_string[i] != check_string[j]:
        print(f"{check_string} is Not a Palindrome")
        break
    i += 1
    j -= 1
else:
    print(f"{check_string} is a Palindrome")
````
