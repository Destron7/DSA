```python
# Printing Numbers using Recursion in descending order
def printNum(num):
    if num > 0:
        print(num)
        printNum(num - 1)
        

printNum(5)
```

```python
def printNumDescModified(num):
    print(num)
    if num == 1:
        return
    printNumDescModified(num - 1)

printNumDescModified(5)
```

```python
# Printing Numbers using Recursion in ascending order
def printNumAsc(num):
    if num > 0:
        printNumAsc(num - 1)
        print(num)

printNumAsc(5)
```

```python
def callBoth(num = 5):
    printNum(num)
    printNumAsc(num - 1)

callBoth()
```

```python
# Facotrial of a number using Recursion
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * (factorial(n-1))

factorial(5)
```

```python
# get 2^n using recursion
def get2powerN(num):
    return 2 * get2powerN(num - 1) if num > 0 else 1

get2powerN(5)
```

```python
# Sum of first N numbers

def sumOfNaturalNumbers(num):
    return num + sumOfNaturalNumbers(num - 1) if num > 1 else 1

sumOfNaturalNumbers(-1)
```

```python
# Number of digits in the number

def numOfDigits(num):
    return 1 + numOfDigits(num // 10) if num > 0 else 0

numOfDigits(18098032)
```

```python
# Fibonnaci using recurrsion

# 1 + 1 + 2 + 3 + 5

def getFibonnaci(num):
    if num == 0 or num == 1:
        return 1
    sum = getFibonnaci(num - 1) + getFibonnaci(num - 2)
    return sum
    
getFibonnaci(5)
```

```python
# Adding digits from a array

def addDigitsOfNum(arr):
    if(len(arr) == 1):
        return arr[0]
    return arr[0] + addDigitsOfNum(arr[1:])

addDigitsOfNum([5])
```

```python
def mergeArraysSorting(arr1: list, arr2: list) -> list:
    merged_array = [0,] * ((size1:= len(arr1)) + (size2:= len(arr2)))
    i , j , k = 0 , 0 , 0
    while i < size1 and j < size2:
        if arr1[i] < arr2[j]:
            merged_array[k] = arr1[i]
            i += 1
        else:
            merged_array[k] = arr2[j]
            j += 1
        k += 1

    for num in arr1[i:]:
        merged_array[k] = num
        k += 1

    for num in arr2[j:]:
        merged_array[k] = num
        k += 1

    return merged_array


mergeArraysSorting([2,3,6],[1,4,6,7,8] )
```
