# Patterns

### Approach:

1. <u>Outer Loop</u> : For Number of Rows. -> Runs number of times lines required.
2. <u>Inner Loop</u> : For Number of Columns. -> Iterates for patterns in the column.


#### *
#### **
#### ***
#### ****
#### *****

```python
i = 0
while i <= 5:
    for j in range(i):
        print("*", end="")
    print()
    i += 1
```
